```python
from pyspark.sql import functions as F

df = spark.table("ons.dados.capacidade_geracao")

# Análise descritiva
display(df.describe())

# Contagem de nulos por coluna
display(
  df.select([
    F.count(
      F.when(
        F.col(c).isNull(),
        c
      )
    ).alias(c)
    for c in df.columns
  ])
)

# Análise de tipos de dados
display(df.dtypes)
```
