```python
source_dir = "s3a://ons-aws-prod-opendata/dataset/capacidade-geracao"
target_dir = "/Volumes/ons/staging/capacidade_geracao"

files = dbutils.fs.ls(source_dir)
for f in files:
    if f.name.endswith(".parquet"):
        dbutils.fs.cp(
            f.path,
            f"{target_dir}/{f.name}"
        )
```
